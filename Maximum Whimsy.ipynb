***My metric is the obesitry rates in Pittsburgh neighborhoods, the data was collected from US Census data***

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

df = pd.read_csv("obesity.csv")
df.info()
```

^^ **initial dataset info**

```python
obesity_df = df.sort_values(by=['City Neighborhood'])
obesity_df.head(20)
```

^^ **sorted by neighborhood**

```python
simp_df = obesity_df.groupby('City Neighborhood')['2006-2010 estimate of obesity'].mean().sort_values(ascending=True)
simp_df.tail(27)
#simp_df.info()
```

^^ **calculated average obesity rate per neighborhood, sorted from least to greatest**<br>All 27 neighborhoods listed are above the national average obesity rate of 41.9%

```python
simp_df.plot.barh(x='2006-2010 estimate of obesity' , figsize = (10,30))
```

^^ **Graph of obesity data sorted Highest to Lowest**

For this particular metric I found that North Oakland is the best neighborhood becasue the obesity rates in that neighborhood are the lowest, which indicates a healthier population. While this may be the best neighborhood initially, when placed in the context of the other two metrics, the best neighborhood will change because our verarching metric concerns obese individuals.
